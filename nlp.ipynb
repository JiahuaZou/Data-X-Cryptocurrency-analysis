```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
import re
from string import punctuation
```

```python
df = pd.read_csv("paradigm_data/articles1000")
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
df.isnull().sum()
```

```python
def remove_punct(text):
    text_nopunct = "".join([char for char in text if char not in punctuation])
    return(text_nopunct)

df['cleaned_description'] = df['description'].apply(lambda x: remove_punct(x))

df.head()
```

```python
def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens

df['tokenized_cleaned_description'] = df['cleaned_description'].apply(lambda x: tokenize(x.lower()))
df.head()
```

```python
def remove_stopwords(tokenized_list):
    text = [word for word in tokenized_list if word not in stopwords.words('english')]
    return text

df['removedstop_tokenized_cleaned_description'] = df['tokenized_cleaned_description'].apply(lambda x: remove_stopwords(x))

df.head()
```

```python
nltk.download('wordnet')
lemmatizer = nltk.stem.WordNetLemmatizer()
ps = nltk.PorterStemmer()
```

```python
def lemmatizing(tokenized_text):
    text = [lemmatizer.lemmatize(word) for word in tokenized_text]
    return text

df['lemmatized_removedstop_tokenized_cleaned_description'] = df['removedstop_tokenized_cleaned_description'].apply(lambda x: lemmatizing(x))

df.head(5)
```

```python
def stemming(tokenized_text):
    text = [ps.stem(word) for word in tokenized_text]
    return text

df['stemmed_removedstop_tokenized_cleaned_description'] = df['removedstop_tokenized_cleaned_description'].apply(lambda x: stemming(x))

df.head()
```
